```python
df['Season'].value_counts().plot(kind='bar')
plt.xlabel('Number of Purchases')
plt.title('Purchases by Season')
plt.ylabel('Season')
plt.show()
```

```python
df['Category'].value_counts().plot(kind='pie')
plt.xlabel('Number of Purchases')
plt.title('Purchases by Category')
plt.ylabel('Category')
plt.show()
```

```python
df.boxplot(column='Review Rating', by='Gender')
plt.title('Review Rating by Gender')
plt.suptitle('')
plt.show()
```

```python
df['Purchase Amount (USD)'].plot(kind='hist', bins=30)
plt.title('Histogram of Purchase Amount')
plt.xlabel('Purchase Amount (USD)')
plt.ylabel('Frequency')
plt.show()
```

```python
df.groupby('Color')['Review Rating'].mean().plot(kind='bar')
plt.title('Average Review Rating by Color')
plt.xlabel('Color')
plt.ylabel('Average Review Rating')
plt.show()
```

```python
df.groupby('Payment Method')['Purchase Amount (USD)'].sum().plot(kind='pie')
plt.title('Total Purchase Amount by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Total Purchase Amount (USD)')
plt.show()
```

```python
df.boxplot(column='Purchase Amount (USD)', by='Frequency of Purchases')
plt.title('Purchase Amount by Frequency of Purchases')
plt.xlabel('Frequency of Purchases')
plt.ylabel('Purchase Amount (USD)')
plt.suptitle('')
plt.show()
```

```python
df.groupby('Season')['Purchase Amount (USD)'].max().plot(kind='pie')
plt.xlabel('Season')
plt.ylabel('Total Purchase Amount (USD)')
plt.title('Total Purchase Amount by Season')
plt.show()
```
